```python
import pandas as pd
import numpy as np

URL = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx'

#data load
df1 = pd.read_excel(URL, sheet_name=0)
df2 = pd.read_excel(URL, sheet_name=1)

#transpose
df1_T = df1.T
df2_T = df2.T

#data dimension / size
df1_D = df1_T.shape
df2_D = df2_T.shape
df1_S = df1_T.size
df2_S = df2_T.size

#data parsing(NaN, NA -> 0)
#print(df1_T.isnull().value_counts())
#print(df2_T.isnull().value_counts())
df1_T.fillna(0)
df2_T.fillna(0)

#Export
df1_T.to_csv('df1.csv', sep=',')
df2_T.to_csv('df2.csv', sep=',')
```
